API stands for Application programming interface<br>
APIs are a way for applications to talk to eachother<br>
In python we use the requests package to handle the connection<br>
<small><strong>Note: It is requests with an s not request(this is a different package)</strong></small><br>
Most API's communicate with data in the form of JSON<br>
JSON stands for JavaScript Object Notation

```python
#import the packages we need:
import requests, json
```

```python
url = 'https://ergast.com/api/f1/2008/driverStandings.json'
```

```python
#response object
response = requests.get(url)
print(response)
print(type(response))
```

```python
#check and make sure we got a successful response from the API
response.ok
```

```python
#view the response as a JSON
response.json()
```

```python
data = response.json()
data
```

```python
# We only want the Driver Standings
x = data['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']
```

```python
def get_top(data):
    new_data = []
    for d in data:
        driver = {}
        d_name = d['Driver']['familyName']
        
        driver[d_name] = {
            'first_name' : d['Driver']['givenName'],
            'last_name' : d['Driver']['familyName'],
            'position' : d['position'], 
            'nation': d['Driver']['nationality'],
            'team': d['Constructors'][0]['constructorId']
        }
        new_data.append(driver)
    return new_data
```

```python
get_top(x)
```

```python
def driver_results(year, round):
    url = f'https://ergast.com/api/f1/{year}/{round}/results.json'
    response = requests.get(url)
    data = response.json()['MRData']['RaceTable']['Races'][0]['Results']
    for d in data :
        print(f"finishing in pos {d['position']} is {d['Driver']['driverId']}")
```

```python
driver_results(2008, 5)
```

```python
import requests, json

def pokeapi(pokemon):
    url =f'https://pokeapi.co/api/v2/pokemon/{pokemon}'
    x = requests.get(url)
    data = x.json()
    
    poke = {
            'ability' : data['abilities'][0]['ability']['name'],
            'name' : data['name'],
            'sprites' : data['sprites']['front_shiny'],
            'hp_base_experience' : data['stats'][0]['base_stat'],
            'attack_base_state' : data['stats'][1]['base_stat'],
            'defense_base_stat' : data['stats'][2]['base_stat']}
        
        
    
    return poke
        

    
pikachu = pokeapi('pikachu')
print(pikachu)
ditto = pokeapi('ditto')
print(ditto)
charizard = pokeapi('charizard')
print(charizard)
eevee = pokeapi('eevee')
print(eevee)
snorlax = pokeapi('snorlax')
print(snorlax)
```
